<h1 style="color:green;text-align:center">Matplotlib: Pie Charts</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

<h2 style="color:purple;text-align:left">Draw pie chart to track down home expenses</h2>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals , labels = exp_labels)
plt.show()
```

<h3 style="color:purple;text-align:left">Show percentages for every pie. Specify radius to increase chart size</h3>

```python

plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%' , radius = 0.8)
plt.show()
```

Explode

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=0.8,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=270)
plt.savefig("piechart.png" , bbox_inches = "tight" , pad_inches = 2)
plt.show()
```
