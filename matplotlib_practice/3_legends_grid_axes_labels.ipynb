<h1 style="color:green; text-align:center;">Matplotlib: Axes Labels, Legend, Grid</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

<h2 style = "color:green; text-align:center;">Axes labels and chart title</h2>

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')


plt.plot(days, max_t, label="max" , color = "black")
plt.plot(days, min_t, label="min" , color = "brown")
plt.plot(days, avg_t, label="average" , color = "pink")
```

<h3 style ="color:green; text-align:center;">Legend</h3>

```python
# Show legend
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label="max temp" , color = "black" , marker = "+" , linestyle = "--" , markersize = 10 , linewidth = 0.5)
plt.plot(days, min_t, label="min temp")
plt.plot(days, avg_t, label="avg temp")

plt.legend(loc = "best")
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label="max temp" , color = "black" , marker = "+" , linestyle = "--" , markersize = 10 , linewidth = 0.5)
plt.plot(days, min_t, label="min temp")
plt.plot(days, avg_t, label="avg temp")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

<h4 style = "color:green;text-align:center">Grid</h4>

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days, max_t, label="max temp" , color = "black" , marker = "+" , linestyle = "--" , markersize = 10 , linewidth = 0.5)
plt.plot(days, min_t, label="min temp")
plt.plot(days, avg_t, label="avg temp")

plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.grid()
```
