<h1 style = "color:blue;text-align:center">Matplotlib: Bar Chart</h1>

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue, color = ['blue','orange','grey','green'])

plt.ylabel("Revenue(Billion)")
plt.xlabel("Companies")
plt.title("US Technology Stocks")
plt.grid(axis='y', linestyle='--', alpha=0.5 , color = "pink")
```

<h2 style = "color:blue;text-align:left">Multiple Bars showing revenue and profit of major US tech companies</h2>

```python
companies = ['Google', 'Amazon', 'Microsoft', 'Apple']
revenue = [282, 514, 212, 394]
profit  = [76, 11, 72, 99]

xpos = np.arange(len(companies))

bar_width = 0.4

plt.figure(figsize=(8,5))

plt.bar(xpos-bar_width/2, revenue, width=bar_width, label='Revenue' , color = 'green')
plt.bar(xpos+bar_width/2, profit, width=bar_width, label='Profit' , color = 'black')
plt.xticks(xpos,companies)

plt.ylabel('Amount (Billion $)')
plt.title('Revenue vs Profit')
plt.legend()
plt.grid(axis = "y" , linestyle = "--" , alpha = 0.5 , color = "pink")
```

```python
companies = ['Google', 'Amazon', 'Microsoft', 'Apple']
revenue = [282, 514, 212, 394]
profit  = [76, 11, 72, 99]

ypos = np.arange(len(companies))

bar_width = 0.4


plt.barh(ypos-bar_width/2 , revenue , label = "REVENUE" , height = bar_width , color = 'lightyellow')
plt.barh(ypos+bar_width/2 , profit , label = "PROFIT" , height = bar_width , color = 'black')
plt.yticks(ypos,companies)
plt.xlabel('Amount (Billion $)')
plt.title('Revenue vs Profit')
plt.legend()
plt.grid(axis = 'x' , linestyle = "--" , alpha = 0.5 , color = "pink")
```
